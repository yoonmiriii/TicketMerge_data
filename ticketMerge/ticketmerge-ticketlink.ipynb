* 'URL' 결측치에 기본 프로필 이미지 넣기 <br>
* 모든 사이트의 공연데이터를 하나로 합치기 <br>

```python
import pandas as pd
import numpy as np
```

```python
df_link = pd.read_csv('../ticketMerge/티켓머지_일정데이터/ticketLink_data.csv')
```

```python
df_link.head(2)
```

```python
# 결측치 확인하기
```

```python
df_link.isna().sum()
```

```python
# 공연기간을 시작일/종료일 두 컬럼으로 나누기
```

```python
dt_link = pd.DataFrame(df_link['공연기간']) 
```

```python
dt_link
```

```python
# "-"를 기준으로 공연기간을 분리하여 새로운 컬럼 추가
```

```python
dt_link = dt_link[['시작일', '종료일']] = dt_link['공연기간'].str.split('-', expand=True)
```

```python
dt_link
```

```python
# 필요 없어진 컬럼 삭제
dt_link = dt_link.drop(columns=[0,1])
```

```python
dt_link
```

```python
# 시작일과 종료일의 결측치 확인하기
# 종료일에 결측치가 있을 경우, 시작일 값으로 채우기 위해서
```

```python
dt_link.isna().sum()
```

```python
# dt_link 데이터 프레임을 원래 데이터프레임에 합치기
```

```python
df_link = pd.concat([df_link, dt_link], axis=1)
```

```python
df_link
```

```python
# 공연기간 컬럼 삭제하기
```

```python
df_link = df_link.drop(['공연기간'], axis=1)
```

```python
df_link.head(1)
```

```python
df_link.isna().sum()
```

```python
# 종료일을 'date' 데이터 타입으로 변환하고 현재일 기준으로, 종료된 공연 삭제하기
```

```python
# '종료일_date' 컬럼 추가하기
```

```python
df_link['종료일_date'] = pd.to_datetime(df_link['종료일'])
```

```python
import datetime as dt
```

```python
dt.datetime.now()
```

```python
current = dt.datetime.now()
```

```python
df_link = df_link[df_link['종료일_date'] >= current]
```

```python
df_link['종료일_date'] >= current
```

```python
# 캐스팅 리스트에서 가수 이름을 , 로 구별하여 나열하기
# [ , ] , '  문자 삭제하기
```

```python
df_link['캐스팅 리스트'] = df_link['캐스팅 리스트'].str.replace('[', '').str.replace(']', '').str.replace("'", "")
```

```python
df_link.to_csv('../ticketMerge/df_link.csv')
```
