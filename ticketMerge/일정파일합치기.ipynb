##### 요구사항
* 출연가수 명단만큼 컬럼 추가하여, 한 컬럼에 한 팀만 넣어주기

```python
import pandas as pd
import pandas as np
```

```python
df1 = pd.read_csv('../ticketMerge/df_inpark.csv')
```

```python
df2 = pd.read_csv('../ticketMerge/df_melon.csv')
```

```python
df3 = pd.read_csv('../ticketMerge/df_link.csv')
```

```python
df4 = pd.read_csv('../ticketMerge/df_yes.csv')
```

```python
# 4개의 파일을 하나로 합치기
```

```python
df = pd.concat([df1, df2, df3, df4], ignore_index=True)
```

```python
df.head(2)
```

```python
df.drop(['Unnamed: 0'], axis=1, inplace=True)
```

```python
df
```

```python
df.isna().sum()
```

```python
# 상세 포스터 URL 결측치에 기본이미지 넣어주기
```

```python
df['상세 포스터 URL'] = df['상세 포스터 URL'].fillna('https://i.namu.wiki/i/GFw2SMaGiqbS3yeTWO08FD5Df8LKk-xB2ckQ0MtDjdoryMCJtsDap9msW17NVgbTM4432kao4DkEGsgwhs_In0zw9vMOldvNbYoc-n1Ng4XgxBVZXNZz33WScPw6zCJY5XAVrdsAFGDx24HN_nu3oQ.webp')
```

```python
df.isna().sum()
```

```python
# 캐스팅 리스트의 쉼표로 구분된 가수 만큼 컬럼을 확장해서, 각 컬럼에 한 팀씩 넣어주기
```

```python
artists = df['캐스팅 리스트'].str.split(', ', expand=True)
```

```python
# 새로운 컬럼 이름을 생성(가수1, 가수2, 가수3...)
```

```python
new_col = [f'가수{i+1}' for i in range(artists.shape[1])]
```

```python
# 기존 데이터프레임에 새로운 컬럼 추가
```

```python
df[new_col] = artists
```
